```python
import matplotlib.pyplot as plt
import numpy as np

# Data
metrics = ['Precision', 'Recall', 'F1-Score', 'Accuracy']
models = ['KNN', 'LR', 'CatBoost',  'GBoost','XGBoost',  'Decision_Tree', 'Random_Forest', 'Bagging_catboost', 'Majority_voting', 'Stacking']
data = [
    [0.2, 0.09, 0.125, 0.77],
    [ 0.629,1, 0.77, 0.895],
    [0.84,0.72,0.78,0.927],
    [0.8,0.73,0.76,0.92],
    [0.84,0.72,0.78,0.93],
    [0.64,0.72,0.68,0.87],
    [0.875,0.64,0.74,0.91],
    [0.89,0.77,0.83,0.94],
    [0.84,0.72,0.78,0.92],
    [0.875,0.64,0.74,0.92]
]

# Set up positions for bars on X-axis
bar_width = 0.09
bar_positions_q1 = np.arange(len(metrics)) - 1 * bar_width
bar_positions_q2 = np.arange(len(metrics)) - 1 * bar_width
bar_positions_q3 = np.arange(len(metrics))
bar_positions_q4 = np.arange(len(metrics)) + 1 * bar_width
bar_positions_q5 = np.arange(len(metrics)) + 1 * bar_width
bar_positions_q6 = np.arange(len(metrics)) + 1 * bar_width

bar_positions_q7 = np.arange(len(metrics)) + 1 * bar_width
bar_positions_q8 = np.arange(len(metrics)) + 1 * bar_width
bar_positions_q9 = np.arange(len(metrics)) + 1 * bar_width
bar_positions_q10 = np.arange(len(metrics)) + 1 * bar_width

plt.figure(figsize=(26, 10))

model_colors = [ '#DEB887','#CDC0B0','#8B8878', '#CDAA7D','#8B7355','#9C661F','#CDAA7D', '#CDC0B0', '#EED5B7', '#DEB887']


for i, (data, color) in enumerate(zip(data, model_colors)):
    bars = plt.bar(
        bar_positions_q1 + i * (bar_width+0.003),
        data,
        width=bar_width,
        label=models[i],
        color = color
    )
    # Add values above the bars
    for bar in bars:
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() ,
                 f'{bar.get_height():.2f}', ha='center', va='bottom')

# Add labels and title
plt.xlabel('Evaluation metrics', fontweight='bold', fontsize=13)
plt.title('Performance Comparision (class_weight)', fontweight='bold', fontsize=15 )
plt.ylabel('Evaluation metrics Value', fontweight='bold', fontsize=13)

# Set X-axis ticks and labels
plt.xticks(bar_positions_q5+0.3, metrics, fontsize=12)
plt.yticks(np.arange(0.1, 1.1, 0.1))

# Add legend inside the plot
plt.legend(loc='upper center', bbox_to_anchor=(0.49, -0.098), ncol=len(models),  fontsize=13)
plt.savefig('smote.pdf', bbox_inches='tight',dpi=800)
# Show the graph
plt.show()


```
